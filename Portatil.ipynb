# GENERAMOS URL DINAMICA

```python
url_base='https://www.alkosto.com/computadores-tablet/computadores-portatiles/c/BI_104_ALKOS?page=1&pageSize=25&sort=relevance'
```

```python
conteo=list(range(1,6))
```

```python
url_din=[]

for i in range(len(conteo)):
  
  url_din.append('https://www.alkosto.com/computadores-tablet/computadores-portatiles/c/BI_104_ALKOS?page='+str(conteo[i])+'&pageSize=25&sort=relevance')
```

```python
tiempo=[10,3,5,7,9,4]
```

```python
random.choice(tiempo)
```

```python
computadores_finales=[]

for i in tqdm(range(len(url_din))):

    peti=requests.get(url_din[i])
    soup=BeautifulSoup(peti.text,'html.parser')
    precio_produto=soup.find_all('div', attrs={'class':'product__price--discounts'})#PRECIOS DE PORTATIL
    descripcion_produto=soup.find_all('div', attrs={'class':'product__information--specifications__block'})
    nombre_portatil=soup.find_all('a', attrs={'class':'js-product-click-datalayer'})
                                
    precio_anterior=[]
    precio_hoy=[]
    espe1=[]
    espe2=[]
    espe3=[]
    espe4=[]
    vespe1=[]
    vespe2=[]
    vespe3=[]
    vespe4=[]

    for j in tqdm(range(len(precio_produto))):


      try :
        precio_hoy.append(precio_produto[j].text.split('$')[2].split()[0])  
      except :
        precio_hoy.append(precio_produto[j].text.split('$')[1].strip())
      
      precio_anterior.append(precio_produto[j].text.split('$')[1].strip())
      espe1.append(descripcion_produto[j].find_all('div', attrs={'class':'item--key'})[0].text)
      vespe1.append(descripcion_produto[j].find_all('div', attrs={'class':'item--value'})[0].text) 
      espe2.append(descripcion_produto[j].find_all('div', attrs={'class':'item--key'})[1].text)
      vespe2.append(descripcion_produto[j].find_all('div', attrs={'class':'item--value'})[1].text)   
      espe3.append(descripcion_produto[j].find_all('div', attrs={'class':'item--key'})[2].text)
      vespe3.append(descripcion_produto[j].find_all('div', attrs={'class':'item--value'})[2].text) 
      espe4.append(descripcion_produto[j].find_all('div', attrs={'class':'item--key'})[3].text)
      vespe4.append(descripcion_produto[j].find_all('div', attrs={'class':'item--value'})[3].text) 
      
    nombre=[]

      # Nombre del portatil
    for k in range(len(nombre_portatil)):

        if k%3==0:

          nombre.append(nombre_portatil[k]['title'])
        else :
          pass
      
      #Data de los portatiles
    data={
            
            'Nombre':nombre, 
            'Precio Oferta': precio_hoy,
            'Precio Anterior': precio_anterior,
            espe1[0]:vespe1,
            espe2[0]:vespe2,
            espe3[0]:vespe3,
            espe4[0]:vespe4,

        }

    computadores=pd.DataFrame(data)
    computadores_finales.append(computadores)
    random.choice(tiempo) # componente de aleatorio
```

```python
len(computadores_finales)
```

```python
df_final=pd.concat(computadores_finales,ignore_index=True)
```
